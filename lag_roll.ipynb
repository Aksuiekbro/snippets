–ö–æ–Ω–µ—á–Ω–æ! –ù–∏–∂–µ ‚Äî **—á—ë—Ç–∫–∏–π –∏ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –ø—Ä–∏–º–µ—Ä** –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è `rolling` –∏ `lag` (—á–µ—Ä–µ–∑ `.shift()`) –Ω–∞ pandas DataFrame, –∫–∞–∫ —ç—Ç–æ —á–∞—Å—Ç–æ –¥–µ–ª–∞—é—Ç –≤ **–≤—Ä–µ–º–µ–Ω–Ω—ã—Ö —Ä—è–¥–∞—Ö / —Ç–∞–±–ª–∏—á–Ω—ã—Ö —Ñ–∏—á–∞—Ö**.

---

## ‚úÖ –ü—Ä–∏–º–µ—Ä: `rolling` –∏ `lag` –¥–ª—è –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö —Ä—è–¥–æ–≤

### üîß –î–∞–Ω–Ω—ã–µ:

```python
import pandas as pd

# –ü—Ä–∏–º–µ—Ä –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞: –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –¥–Ω—è–º
df = pd.DataFrame({
    'date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    'sales': [100, 110, 90, 95, 130, 150, 160, 170, 180, 190]
})

print(df)
```

---

### üìç 1. `lag` ‚Äî —Å–¥–≤–∏–≥–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ 1 –¥–µ–Ω—å –Ω–∞–∑–∞–¥

```python
df['lag_1'] = df['sales'].shift(1)
```

---

### üìç 2. `rolling` ‚Äî —Å—Ä–µ–¥–Ω–µ–µ –∑–∞ 3 –¥–Ω—è

```python
df['rolling_mean_3'] = df['sales'].rolling(window=3).mean()
```

| 2023-01-02 | 110   | NaN              |
| 2023-01-03 | 90    | 100.0            |

| 2023-01-05 | 130   | 105.0            |

---

### üìç 3. –ß–∞—Å—Ç–æ –∏—Å–ø–æ–ª—å–∑—É—é—Ç –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏:

```python
# –ù–∞–ø—Ä–∏–º–µ—Ä, —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É —Ç–µ–∫—É—â–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –∏ –ø—Ä–µ–¥—ã–¥—É—â–∏–º
df['diff_1'] = df['sales'] - df['lag_1']

# –ò–ª–∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –∑–∞ –æ–∫–Ω–æ
df['rolling_std_3'] = df['sales'].rolling(window=3).std()
```

---

## üß† –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –≤ ML

–¢–∞–∫–∏–µ —Ñ–∏—á–∏ (`lag`, `rolling mean`, `rolling std`, `rolling min/max`) —á–∞—Å—Ç–æ:

* –¥–æ–±–∞–≤–ª—è—é—Ç –¥–ª—è –º–æ–¥–µ–ª–µ–π –≤ time-series (LightGBM, CatBoost)
* –ø–æ–∑–≤–æ–ª—è—é—Ç —É—á–∏—Ç—ã–≤–∞—Ç—å –¥–∏–Ω–∞–º–∏–∫—É (—Ä–æ—Å—Ç, –ø–∞–¥–µ–Ω–∏–µ, —Ç—Ä–µ–Ω–¥)
* —Ö–æ—Ä–æ—à–æ —Ä–∞–±–æ—Ç–∞—é—Ç –¥–∞–∂–µ –≤ –æ–±—ã—á–Ω—ã—Ö —Ä–µ–≥—Ä–µ—Å—Å–∏—è—Ö

---

–•–æ—á–µ—à—å –ø—Ä–∏–º–µ—Ä, –∫–∞–∫ —ç—Ç–∏ —Ñ–∏—á–∏ –ø–æ–¥–∞—é—Ç—Å—è –≤ LightGBM –∏–ª–∏ CatBoost –≤ time-series –∑–∞–¥–∞—á–µ?


–û—Ç–ª–∏—á–Ω–æ, –ø–æ–∫–∞–∂—É –ø–æ–ª–Ω—ã–π –ø—Ä–∏–º–µ—Ä: **`rolling` + `lag` + —É–¥–∞–ª–µ–Ω–∏–µ NaN**, –≤—Å—ë –∫–∞–∫ –Ω—É–∂–Ω–æ –¥–ª—è –ø–æ–¥–∞—á–∏ –≤ ML-–º–æ–¥–µ–ª—å.

---

## ‚úÖ –ü—Ä–∏–º–µ—Ä: –ª–∞–≥–∏, —Å–∫–æ–ª—å–∑—è—â–µ–µ –æ–∫–Ω–æ, –æ—á–∏—Å—Ç–∫–∞

```python
import pandas as pd

# –ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –¥–Ω—è–º
df = pd.DataFrame({
    'date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    'sales': [100, 110, 90, 95, 130, 150, 160, 170, 180, 190]
})

# –õ–∞–≥ –Ω–∞ 1 –¥–µ–Ω—å –Ω–∞–∑–∞–¥
df['lag_1'] = df['sales'].shift(1)

# –°–∫–æ–ª—å–∑—è—â–µ–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–∞ 3 –¥–Ω—è
df['rolling_mean_3'] = df['sales'].rolling(window=3).mean()

# –°–∫–æ–ª—å–∑—è—â–µ–µ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
df['rolling_std_3'] = df['sales'].rolling(window=3).std()

# –†–∞–∑–Ω–∏—Ü–∞ —Å –ø—Ä–µ–¥—ã–¥—É—â–∏–º –¥–Ω—ë–º
df['diff_1'] = df['sales'] - df['lag_1']

# –£–¥–∞–ª—è–µ–º —Å—Ç—Ä–æ–∫–∏ —Å NaN (–≤–æ–∑–Ω–∏–∫–∞—é—Ç –≤ –Ω–∞—á–∞–ª–µ –ø—Ä–∏ lag/rolling)
df_clean = df.dropna().reset_index(drop=True)

print(df_clean)
```

## üìå –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ:

* `dropna()` –Ω—É–∂–µ–Ω, –ø–æ—Ç–æ–º—É —á—Ç–æ `shift()` –∏ `rolling()` —Å–æ–∑–¥–∞—é—Ç NaN –≤ –Ω–∞—á–∞–ª–µ.
* –ü–µ—Ä–µ–¥ –ø–æ–¥–∞—á–µ–π –≤ –º–æ–¥–µ–ª—å (`XGBoost`, `CatBoost`, `LightGBM`) **–æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ —Å–±—Ä–æ—Å—å –∏–Ω–¥–µ–∫—Å**, –∏–Ω–∞—á–µ –±—É–¥–µ—Ç –º—É—Å–æ—Ä –≤ –∏–Ω–¥–µ–∫—Å–∞—Ö.

---

–•–æ—á–µ—à—å, —è –ø–æ–∫–∞–∂—É, –∫–∞–∫ —ç—Ç—É —Ç–∞–±–ª–∏—Ü—É –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤ –º–æ–¥–µ–ª–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, LightGBM) –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è `sales`?


–û—Ç–ª–∏—á–Ω–æ, –ø–æ–∫–∞–∂—É –ø–æ–ª–Ω—ã–π –ø—Ä–∏–º–µ—Ä: **`rolling` + `lag` + —É–¥–∞–ª–µ–Ω–∏–µ NaN**, –≤—Å—ë –∫–∞–∫ –Ω—É–∂–Ω–æ –¥–ª—è –ø–æ–¥–∞—á–∏ –≤ ML-–º–æ–¥–µ–ª—å.

---

## ‚úÖ –ü—Ä–∏–º–µ—Ä: –ª–∞–≥–∏, —Å–∫–æ–ª—å–∑—è—â–µ–µ –æ–∫–Ω–æ, –æ—á–∏—Å—Ç–∫–∞

```python
import pandas as pd

# –ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –¥–Ω—è–º
df = pd.DataFrame({
    'date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    'sales': [100, 110, 90, 95, 130, 150, 160, 170, 180, 190]
})

# –õ–∞–≥ –Ω–∞ 1 –¥–µ–Ω—å –Ω–∞–∑–∞–¥
df['lag_1'] = df['sales'].shift(1)

# –°–∫–æ–ª—å–∑—è—â–µ–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–∞ 3 –¥–Ω—è
df['rolling_mean_3'] = df['sales'].rolling(window=3).mean()

# –°–∫–æ–ª—å–∑—è—â–µ–µ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
df['rolling_std_3'] = df['sales'].rolling(window=3).std()

# –†–∞–∑–Ω–∏—Ü–∞ —Å –ø—Ä–µ–¥—ã–¥—É—â–∏–º –¥–Ω—ë–º
df['diff_1'] = df['sales'] - df['lag_1']

# –£–¥–∞–ª—è–µ–º —Å—Ç—Ä–æ–∫–∏ —Å NaN (–≤–æ–∑–Ω–∏–∫–∞—é—Ç –≤ –Ω–∞—á–∞–ª–µ –ø—Ä–∏ lag/rolling)
df_clean = df.dropna().reset_index(drop=True)

print(df_clean)
```

## üìå –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ:

* `dropna()` –Ω—É–∂–µ–Ω, –ø–æ—Ç–æ–º—É —á—Ç–æ `shift()` –∏ `rolling()` —Å–æ–∑–¥–∞—é—Ç NaN –≤ –Ω–∞—á–∞–ª–µ.
* –ü–µ—Ä–µ–¥ –ø–æ–¥–∞—á–µ–π –≤ –º–æ–¥–µ–ª—å (`XGBoost`, `CatBoost`, `LightGBM`) **–æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ —Å–±—Ä–æ—Å—å –∏–Ω–¥–µ–∫—Å**, –∏–Ω–∞—á–µ –±—É–¥–µ—Ç –º—É—Å–æ—Ä –≤ –∏–Ω–¥–µ–∫—Å–∞—Ö.

---

–•–æ—á–µ—à—å, —è –ø–æ–∫–∞–∂—É, –∫–∞–∫ —ç—Ç—É —Ç–∞–±–ª–∏—Ü—É –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤ –º–æ–¥–µ–ª–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, LightGBM) –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è `sales`?


–û—Ç–ª–∏—á–Ω—ã–π –≤–æ–ø—Ä–æ—Å. –í–º–µ—Å—Ç–æ —É–¥–∞–ª–µ–Ω–∏—è —Å—Ç—Ä–æ–∫ —Å `NaN` (`dropna()`), –º–æ–∂–Ω–æ **–∑–∞–ø–æ–ª–Ω–∏—Ç—å –∏—Ö –≤—Ä—É—á–Ω—É—é** ‚Äî —ç—Ç–æ —á–∞—Å—Ç–æ –¥–µ–ª–∞—é—Ç –≤ ML, –æ—Å–æ–±–µ–Ω–Ω–æ –µ—Å–ª–∏ —Ç—ã –Ω–µ —Ö–æ—á–µ—à—å —Ç–µ—Ä—è—Ç—å –¥–∞–Ω–Ω—ã–µ.

–í–æ—Ç –∫–∞–∫ —ç—Ç–æ –¥–µ–ª–∞–µ—Ç—Å—è:

---

## ‚úÖ –°–ø–æ—Å–æ–±—ã –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è `NaN`

```python
import pandas as pd
import numpy as np

# –ü—Ä–∏–º–µ—Ä —Å NaN –ø–æ—Å–ª–µ rolling/shift
df = pd.DataFrame({
    'sales': [100, 110, 90, 95, 130, 150, 160, 170, 180, 190]
})

df['lag_1'] = df['sales'].shift(1)
df['rolling_mean_3'] = df['sales'].rolling(3).mean()
df['rolling_std_3'] = df['sales'].rolling(3).std()
```

---

### üìç 1. –ó–∞–ø–æ–ª–Ω–∏—Ç—å –Ω—É–ª—è–º–∏ (`fillna(0)`)

```python
df.fillna(0, inplace=True)
```

‚úÖ –ü—Ä–æ—Å—Ç–æ –∏ –±–µ–∑–æ–ø–∞—Å–Ω–æ, –µ—Å–ª–∏ `0` ‚Äî –¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è –≤—ã—Ä—É—á–∫–∏).

---

### üìç 2. –ó–∞–ø–æ–ª–Ω–∏—Ç—å –ø—Ä–µ–¥—ã–¥—É—â–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ–º (`ffill`)

```python
df.fillna(method='ffill', inplace=True)  # forward fill
```

‚úÖ –ü–æ–¥—Ö–æ–¥–∏—Ç, –µ—Å–ª–∏ –∑–Ω–∞—á–µ–Ω–∏—è –ø–ª–∞–≤–Ω—ã–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –≤—Ä–µ–º—è).

---

### üìç 3. –ó–∞–ø–æ–ª–Ω–∏—Ç—å —Å—Ä–µ–¥–Ω–∏–º/–º–µ–¥–∏–∞–Ω–æ–π

```python
df['rolling_mean_3'].fillna(df['rolling_mean_3'].mean(), inplace=True)
df['lag_1'].fillna(df['lag_1'].median(), inplace=True)
```

‚úÖ –ß–∞—Å—Ç–æ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–ª—è —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø—Ä–∏ –æ–±—É—á–µ–Ω–∏–∏ –º–æ–¥–µ–ª–µ–π.

---

### üìç 4. –ö–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø–æ–¥—Ö–æ–¥ (–≥–∏–±–∫–∏–π)

```python
df['lag_1'] = df['lag_1'].fillna(0)
df['rolling_mean_3'] = df['rolling_mean_3'].fillna(method='ffill').fillna(0)
df['rolling_std_3'] = df['rolling_std_3'].fillna(0)
```

| --------------- | ------------------------------ |

| `ffill`         | –í—Ä–µ–º–µ–Ω–Ω—ã–µ —Ä—è–¥—ã, smooth data    |
